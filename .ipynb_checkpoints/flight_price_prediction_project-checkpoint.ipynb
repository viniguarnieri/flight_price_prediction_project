```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import category_encoders as ce
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score


# Extract
def show_first_5_lines(data):
    print('                                             Initial Visualization                                                ')
    print('------------------------------------------------------------------------------------------------------------------')
    print('                                                 First 5 lines                                                    ')
    display(data.head())
    print('==================================================================================================================')
    return None


def show_last_5_lines(data):
    print('                                                  Last 5 lines                                                    ')
    display(data.tail())
    print('==================================================================================================================')
    return None


def count_empty_columns(data):
    print('                                            Number of null columns                                                ')
    print(data.isnull().sum())
    print('==================================================================================================================')
    return None


def show_data_types(data):
    print('                                                   Data types                                                     ')
    print(data.dtypes)
    print('==================================================================================================================')
    return None


# Transform
def drop_columns(data):
    data = data.drop(['Unnamed: 0', 'flight'], axis=1)
    return data


def plot_price_statistics (data):
    
    print('                                                     EDA                                                          ')
    print('------------------------------------------------------------------------------------------------------------------')
    
    print('                                                    PRICE                                                         ')
    
    # Get the label column
    label = data['price']

    # Get statistics
    label_min = label.min()
    label_max = label.max()
    label_mean = label.mean()
    label_median = label.median()
    label_mode = label.mode()[0]

    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(label_min,
                                                                                                label_max,
                                                                                                label_mean,
                                                                                                label_median,
                                                                                                label_mode))
        
    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(2, 1, figsize=(7,9))

    # Plot the histogram
    ax[0].hist(label)

    # Add lines for the min, max, mean, median, and mode
    ax[0].axvline(label_min, color='gray', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_max, color='gray', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_mean, color='cyan', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_median, color='magenta', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_mode, color='yellow', linestyle='dashed', linewidth=2)


    # Plot the boxplot
    ax[1].boxplot(label, vert=False)

    # Titles
    for i in range(0,2):
        ax[i].set_title('Data Distribution')
        ax[i].set_xlabel('Price')
        ax[i].set_ylabel('Frequency')

    # Show the figure
    plt.show()
    print('==================================================================================================================')
    
    return data


def plot_numeric_features_correlation(data):
    
    print('                                              Numeric Features                                                    ')
    # Get numeric features
    numeric_features = data.select_dtypes(include=['float64', 'int64']).drop('price', axis=1)
    
    # Plot a histogram for each numeric feature
    for col in numeric_features:
        fig = plt.figure(figsize=(7, 4))
        ax = fig.gca()
        feature = data[col]
        feature.hist(bins=100, ax = ax)
        ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
        ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
        label = data['price']
        correlation = feature.corr(label)
        ax.set_title('price vs ' + col + ' - correlation: ' + str(correlation))
    plt.show()
    print('==================================================================================================================')
    
    return data


def plot_categorical_features_correlation(data):
    
    print('                                            Categorical Features                                                  ')
    # Get categorical features
    categorical_features = data.select_dtypes(include=('object'))

    # Plot a boxplot for each categorical feature
    for col in categorical_features:
        fig = plt.figure(figsize=(7, 4))
        ax = fig.gca()
        data.boxplot(column = 'price', by = col, ax = ax)
        ax.set_title('Label by ' + col)
        ax.set_ylabel("price")
    plt.show()
    print('==================================================================================================================')
    
    return data


def question_a(data):
    
    print('                                          Answering the Questions                                                 ')
    print('------------------------------------------------------------------------------------------------------------------')
    print('a) Does price vary with Airlines?')
    
    # Ensure plots are displayed inline in the notebook
    %matplotlib inline

    # Create the grouping
    label = data[['airline', 'price']].groupby('airline').mean().reset_index()

    # Create a Figure
    fig = plt.figure(figsize=(8,5))

    # Customize the chart
    colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#CCB974', '#64B5CD']
    plt.bar(x=label.airline, height=label.price, color=colors)
    plt.title('Price vs Airlines')
    plt.xlabel('Airlines')
    plt.ylabel('Price')

    # Show the figure
    plt.show()
    
    return data


def question_b(data):
    
    print('b) How is the price affected when tickets are bought in just 1 or 2 days before departure?')
    
    # Create de grouping
    df = data[data['days_left'] <= 2] 
    label = df[['days_left', 'price']].groupby('days_left').mean().reset_index()

    # Create a Figure
    fig = plt.figure(figsize=(7,4))

    # Customize the chart
    colors = ['#4C72B0', '#55A868']
    plt.bar(x=label.days_left, height=label.price, color=colors)
    plt.title('Price vs Days before departure')
    plt.xlabel('Days before departure')
    plt.ylabel('Price')

    # Show the figure
    plt.show()
    
    return data


def question_c(data):
    
    print('c) Does ticket price change based on the departure time and arrival time?')
    
    # Get the label column
    df_1 = data[['departure_time', 'price']].groupby('departure_time').mean().reset_index()
    df_2 = data[['arrival_time', 'price']].groupby('arrival_time').mean().reset_index()

    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(1,2, figsize=(15,5))

    # First plot
    ax[0].bar(x=df_1.departure_time, height=df_1.price, color='orange')
    ax[0].set_title('Price vs Departure_time')
    ax[0].set_xticklabels(df_1.departure_time)

    # Second Plot
    ax[1].bar(x=df_2.arrival_time, height=df_2.price, color='#C44E52')
    ax[1].set_title('Price vs Arrival_time')
    ax[1].set_xticklabels(df_2.arrival_time)

    # Show the figure
    plt.show()
    
    return data


def question_d(data):
    
    print('d) How the price changes with change in Source and Destination?')
    
    # Get the label column
    df_1 = data[['source_city', 'price']].groupby('source_city').mean().reset_index()
    df_2 = data[['destination_city', 'price']].groupby('destination_city').mean().reset_index()

    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(1,2, figsize=(15,5))

    # First plot
    ax[0].bar(x=df_1.source_city, height=df_1.price, color='#64B5CD')
    ax[0].set_title('Price vs Source_city')
    ax[0].set_xticklabels(df_1.source_city)

    # Second Plot
    ax[1].bar(x=df_2.destination_city, height=df_2.price, color='#55A868')
    ax[1].set_title('Price vs Destination_city')
    ax[1].set_xticklabels(df_2.destination_city)

    # Show the figure
    plt.show()
    
    return data


def question_e(data):
    
    print('e) How does the ticket price vary between Economy and Business class?')
    # Create a Figure
    fig = plt.figure(figsize=(8,5))

    # Plot
    class_counts = data['class'].value_counts()
    plt.pie(class_counts, labels=class_counts)
    plt.title('Price vs Class')
    plt.legend(class_counts.keys().tolist())

    # Show the figure
    plt.show()
    print('==================================================================================================================')
    
    return data


def machine_learning(data):
    
    # Normalization
    print('                                                Data Modeling                                                     ')
    print('------------------------------------------------------------------------------------------------------------------')
    print('                                      Categorical Features Normalization                                          ')
    te = ce.TargetEncoder(cols=data.select_dtypes(include='object'))
    data = te.fit_transform(data, data['price'])
    display(data.head())
    print('==================================================================================================================')

    # Scaling
    print('                                                Scaling Data                                                     ')
    # Get a scaler object
    scaler = MinMaxScaler()

    # Apply the scaler
    data = pd.DataFrame(scaler.fit_transform(data))
    
    # Rename the columns
    data.columns = ['airline', 'source_city', 'departure_time', 'stops', 'arrival_time', 'destination_city', 'class', 'duration', 'days_left', 'price']
        
    # Show DataFrame
    display(data.head())
    print('==================================================================================================================')
    
    # Pearson Correlation
    print('                                            Pearson Correlation                                                   ')
    plt.figure(figsize=(25,10))
    cor = data.corr()
    sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
    plt.show()
    print('==================================================================================================================')
    
    # Correlation with target variable price
    print('                                             Relevant Features                                                    ')
    cor_target = abs(cor["price"])
    relevant_features = cor_target
    print(relevant_features)
    print('==================================================================================================================')
    
    # Machine Learning
    print('                                              Machine Learning                                                    ')
    print('------------------------------------------------------------------------------------------------------------------')
    # Target variable
    y = data['price']

    # Input features
    X_data = data.drop(['price'], axis=1)
    
    # Split data 70%-30% into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(X_data, y, test_size=0.30, random_state=0)

    print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
    display(data.head())
    
    
    # Linear Regression
    print('==================================================================================================================')
    print('                                             Linear Regression                                                    ')
    
    # Fit a Linear Regression model on the training set
    lr = LinearRegression().fit(X_train, y_train)

    # Make predictions
    y_pred = lr.predict(X_test)

    # Evaluate the model on the testing data
    lr_mse = mean_squared_error(y_test, y_pred)
    lr_rmse = np.sqrt(lr_mse)
    lr_r2 = r2_score(y_test, y_pred)

    print("Mean squared error:", lr_mse)
    print("RMSE :", lr_rmse)
    print("R-squared:", lr_r2)
    
    
    # Decision Tree Regressor
    print('==================================================================================================================')
    print('                                               Decision Tree                                                      ')
    
    # Model
    dt = DecisionTreeRegressor(max_depth=5, min_samples_split=10)
    
    # Fit a Decision Tree regression model on the training set
    dt.fit(X_train, y_train)
    
    # Make predictions
    y_pred = dt.predict(X_test)

    # Evaluate the model on the testing data
    dt_mse = mean_squared_error(y_test, y_pred)
    dt_rmse = np.sqrt(dt_mse)
    dt_r2 = r2_score(y_test, y_pred)

    print("Mean squared error:", dt_mse)
    print("RMSE :", dt_rmse)
    print("R-squared:", dt_r2)
    
    
    # Random Forest Regressor
    print('==================================================================================================================')
    print('                                               Random Forest                                                      ')
    
    # Model
    rf = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)

    # Fit a Random Forest model on the training set
    rf.fit(X_train, y_train)

    # Make predictions
    y_pred = rf.predict(X_test)

    # Evaluate the model on the testing data
    rf_mse = mean_squared_error(y_test, y_pred)
    rf_rmse = np.sqrt(rf_mse)
    rf_r2 = r2_score(y_test, y_pred)

    print("Mean squared error:", rf_mse)
    print("RMSE :", rf_rmse)
    print("R-squared:", rf_r2)
    
    return None


def data_collect (path):
    # Load dataset
    data = pd.read_csv(path)
    
    show_first_5_lines(data)
    
    show_last_5_lines(data)
    
    count_empty_columns(data)
    
    show_data_types(data)
    
    return data


def data_transform (data):

    data = drop_columns(data)
    print('Dropped columns: "Unnamed: 0" and "flight"')
    display(data.head())
    print('==================================================================================================================')
    
    question_a(data)
    question_b(data)
    question_c(data)
    question_d(data)
    question_e(data)

    plot_price_statistics (data)
  
    plot_numeric_features_correlation(data)
    
    plot_categorical_features_correlation(data)
    
    machine_learning(data)

    return data


if __name__ == '__main__':
    #ETL
    data_raw = data_collect ('Clean_Dataset.csv')
    
    data_processing = data_transform (data_raw)
```
